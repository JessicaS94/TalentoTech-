PRIMER EJERCICIO

```python
#Ejercicio 1 - Calcular el total de ventas por region

print('Ejercicio 1: Calcular el total de ventas por region')
ventas_por_region = df.groupby('Region')['Ventas'].sum()
print(ventas_por_region)
print("\n")
print(df.head(18))
```

```python
#Ejercicio 2: Filtrar el df para mostrar solo las ventas realizadas por david

print('Ejercicio 2: Filtrar el dataframe para solo mostrar las ventas realizadas por David')
ventas_david = df[df['Vendedor'] == 'David']
print(ventas_david)
print("\n")

```

```python
#Ejercicio 3: calcular el promedio de ventas por producto

print('Ejercicio 3: calcular el promedio de ventas por producto')
ventas_por_producto = df.groupby('Producto')['Ventas'].mean()
print(ventas_por_producto)
print("\n")

```

```python
#Ejecicio 4: Ordenar el df por ventas en orden descendente
print('Ejercicio 4: Ordenar el df por ventas en orden descendente')
df_ordenado = df.sort_values(by=['Ventas'], ascending=False)
print(df_ordenado)
```

MANIPULACIÓN DE DATOS

```python
#Ejercicio 5: Filtrar las filas con ventas mayores a 10000

print('Ejercicio 5: Filtrar las filas con ventas mayores a 10000')
filtrar_ventas = df[df['Ventas'] > 10000]
print(filtrar_ventas)
print("\n")
```

```python
#Ejercicio 6: Ordenar el dataframe por ventas en orden descendente

print('Ejercicio 6: Ordenar el dataframe por ventas en orden descendente')
df_ordenado_ascendente = df.sort_values(by=['Ventas'])
print(df_ordenado_ascendente)
print("\n")
```

```python
#Ejercicio 7: eliminar columna "vendedor"

print('Ejercicio 7: eliminar columna "vendedor"')
eliminar_columna = df.drop(columns=['Vendedor'])
print(eliminar_columna)
print("\n")
```
